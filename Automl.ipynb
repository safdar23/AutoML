```python
%%sh
wget -N https://archive.ics.uci.edu/ml/machine-learning-databases/00222/bank-additional.zip
unzip -o bank-additional.zip
```

```python
import pandas as pd
data = pd.read_csv('./bank-additional/bank-additional-full.csv', sep = ";")
pd.set_option('display.max_columns',500)
pd.set_option('display.max_rows',50)
data[:10]
```

```python
import numpy as np
train_data, test_data, _=np.split(data.sample(frac=1,random_state=123), [int(0.95 * len(data)),int(len(data))])
#save to csv file
train_data.to_csv('bank-train.csv', index=False, header=True,sep = ",")
test_data.to_csv('bank-test.csv', index=False, header=True,sep = ",")
```

```python
import sagemaker
prefix ='sagemaker/DEMO-automl'
sess =sagemaker.Session()
uri = sess.upload_data(path='bank-train.csv', key_prefix=prefix)
```
